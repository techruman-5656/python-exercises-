```python
def q1(s):
    reversed_string = s[::-1]
    second_char_string = s[1::2]
    return reversed_string, second_char_string

print(q1("hello world"))
```

```python
def q2(lst):
    return [x**2 for x in lst if x % 2 == 0]

print(q2([1,2,3,4,5,6]))
```

```python
def q3(s):
    freq = {}
    for c in s:
        freq[c] = freq.get(c,0)+1
    return freq

print(q3("banana"))
```

```python
def q4(lst):
    return [i for i,v in enumerate(lst) if v % 2 != 0]

print(q4([1,2,3,4,5,6,7]))
```

```python
def q5(lst,k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

print(q5([1,2,3,4,5],2))
```

```python
def q6(list1,list2):
    return dict(zip(list1,list2))

print(q6(["a","b","c"],[1,2,3]))
```

```python
def q7(lst):
    it = iter(lst)
    try:
        while True:
            print(next(it))
    except StopIteration:
        pass

q7([10,20,30,40])
```

```python
def q8(nested):
    return [item for sublist in nested for item in sublist]

print(q8([[1,2],[3,4],[5,6]]))
```

```python
def q9(s):
    return s == s[::-1]

print(q9("madam"))
print(q9("hello"))
```

```python
def q10(lst):
    first = second = float('-inf')
    for n in lst:
        if n > first:
            second = first
            first = n
        elif first > n > second:
            second = n
    return second

print(q10([10,20,4,45,99]))
```
